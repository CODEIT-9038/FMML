```python
#STRINGS

str = "   Lokesh Mishra   "

print(str.strip()) #removes all whitespaces in the start or at the end

print(str.islower()) #checks weather a string is lowrcase or not

print(str.isupper()) #checks weaher a string is uppercase or not

print(str.lower()) #convert string to lowercase

print(str.upper()) #convert string to uppercase
```

```python
#LIST

l = [1,2,3,4]
l.append(5) #Adds an element at the end of the list
print(l)

l.clear() #Removes all the elements from the list
print(l)

l = [1,2,2,3,2,4]
print(l.count(2)) #print the no of times an element is present in a string

l.pop(3) #removes the element in the specified index
print(l)

l2 = [0]
l2.extend(l) #extend()
print(l2)
```

Question 1 :
Define a function, which takes in two strings A and B. Reverses the first string A, adds it to B, and returns the final string.

```python
def custom(A:str, B:str):
  B += A[::-1]
  return B

A = "Lokesh"
B = "Mishra"

print(custom(A,B))
```

Question 2 :

Given a list having Names, work_hours, and gender, Write a function to print name of the female worker that worked the most hours. Also how much do should she be paid if the pay is $ 20 per hour.

work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]

Answer : the female worker that worked the most hours is Penny and she should be paid 18400

```python
def hardWorker(workers: list):
  name = ""
  max = 0

  for worker in workers:
    if(worker[2] == 'F' and worker[1] >= max):
      max = worker[1]
      name = worker[0]

  print(f"the female worker that worked the most hours is {name} and she should be paid {max*20}")

work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]
hardWorker(work_hours)
```

Question 3 :

a)Create a 5x5 matrix of the following form,

[[1,1]
[2,2]]

i.e. each row is increasing and has repetive elements.

Hint : you can use hstack, vstack etc.

b) find dot product of the matrix with any matrix. 

(Figure out the size/ shape of the matrix)

```python

import numpy as np

def stackedArray(n: int):

  temp = np.vstack([np.hstack([i for j in range(n)]) for i in range(1,n+1)])
  
  return temp

speacialMatrix = stackedArray(5)

finalMatrix = np.dot(speacialMatrix, np.eye(5))

print(finalMatrix.size)
print(finalMatrix.shape)

```
